```
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')
```

```
tweets = pd.read_csv("/content/gdrive/My Drive/WisdumbAI/naval.csv")
```

```
tweets.head(-10)
```

```
tweets['tweet'].astype(str)
tweets['tweet'] = tweets['tweet'].str.replace('@\S+|pic.\S+|http\S+', 'deletethisrow', case=False)
tweets['mentions'] = tweets['mentions'].str.replace('\'\S+', 'deletethisrow', case=False)
tweets['tweet'] = tweets['tweet'].str.replace('\n|#', ' ', case=False)
```

```
naval = tweets[~tweets.tweet.str.contains('deletethisrow', case=False)]
naval = naval[~naval.mentions.str.contains('deletethisrow', case=False)]
```

```
naval = naval['tweet']
```

```
naval.size
```

```
naval.to_csv('naval_clean.csv',index=False)
```
